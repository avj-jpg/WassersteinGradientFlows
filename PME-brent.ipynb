```python
from PME import PME1D, PME2D
import numpy as np
import matplotlib.pyplot as plt
plt.ion()
```

# One dimension

## Sample solution

```python
P = PME1D(beta=0.01, nx=32, ny =32, order=2)
print(P)
P.solve()
P.draw(P.rhoh)
```

## Convergence study

```python
nx, ny = 8, 8
beta = 0.01

labels = [r'$8\times 8$', r'$16\times 16$', r'$32\times 32$', r'$64\times64$']
colors = ['r','g','b','m']
```

### $\alpha = 1.5$

```python
alpha = 1.5
```

$p=1$

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))

nIter = 1000
order = 1
for i in range(4):
    P1 = PME1D(order = order, maxIter = nIter + 1, alpha = alpha, beta = beta, nx = nx * 2 ** i, ny = ny * 2 ** i)
    if i == 0: print(P1)
    P1.solve()
    P1.plotErr(axs, labels[i], colors[i])
    if i > 0:
        print('Order (ST) = ', np.log2(err0/P1.stErr[-1]))
        print('Order (Terminal) = ', np.log2(err1/P1.terminalErr[-1]))
        print("\n")
    err0 = P1.stErr[-1]
    err1 = P1.terminalErr[-1]
    
axs[0].legend()
plt.show()
```
