```python
from PME import PME1DBB
import numpy as np
import matplotlib.pyplot as plt
plt.ion()
```

```python
def plotErr(self, axs, label, color):
    for ax in axs:
        ax.set_xlabel("Iteration count")
    axs[0].set_ylabel("PDHG Error")
    axs[1].set_ylabel("Rho Errors")
    axs[0].semilogy(self.pdhgErr, "-", label = label, color = color)
    axs[1].semilogy(self.stErr, "-", label = label, color = color)
    axs[1].semilogy(self.terminalErr, "--", label = label, color = color)
    axs[1].text(
        0.95, 0.95, 
        " — spacetime error\n --- terminal error", 
        transform=axs[1].transAxes, 
        fontsize=10,
        color="red",
        horizontalalignment='right',
        verticalalignment='top',
        multialignment='left',
        bbox=dict(boxstyle="round,pad=0.3", edgecolor="gray", facecolor="white", alpha=0.5)
)
```

```python
P2 = PME1DBB(sc= .1, alpha=2, beta=.5, nx=64, ny = 32, order=2, xmax=2, maxIter=1000, printNum=50)
P2.solve()
P2.saveVTK("m2")
```

```python
P2.draw(P2.rhoh)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
plotErr(P2,axs,'128x64','red')
```

```python
P3 = PME1DBB(sc= .1, alpha=3, beta=.5, nx=64, ny = 32, order=3, xmax=2, maxIter=2500, printNum=100)
P3.solve()
P3.saveVTK("m3")
```

```python
P3.draw(P3.rhoh)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
plotErr(P3,axs,'','red')
```
