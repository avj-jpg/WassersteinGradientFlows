```python
from FokkerPlanck import FokkerPlanck1D
import matplotlib.pyplot as plt
from ngsolve import *
%matplotlib widget
```

```python
F = FokkerPlanck1D(K=.5,v=5, nx=50, ny=50, nonLinearSolver="brent", brentMin=0, brentMax=6, order=1, beta=.05)
F.draw(F.rho_exact())
```

```python
F.solve()
```

```python
fig, ax = plt.subplots(figsize = (9,6))
F.animate(fig, ax, color='b')
```

```python
fig, ax = plt.subplots(figsize = (9,6))
F.snapshots(fig,ax)
```
